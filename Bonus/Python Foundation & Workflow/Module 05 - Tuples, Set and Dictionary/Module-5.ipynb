```python
# Declaration

A = {1, 2, 3}

print(type(A), A)
```

```python
# Empty Structures

B = []

print(type(B))

C = ()

print(type(C))

S = set()

print(type(S))


```

```python
# Accessing

S = {1, 2, 3, 4, 1, 2}

# Checking if element is there on the set
if (3 in S):
    print("3 is present")
else:
    print("Not present")

sum = 0

for element in S:
    print(element)
    sum += element

print(sum)
```

```python
# Mutable

s = {5, 1, 2, 4, 3}

s.add(10)
s.add(0)

print(s)

s.pop()

print(s)

s.remove(5)

print(s)
```

### Set Mathematical Methods

```python
a = {1, 2, 3}
b = {1, 2, 3, 4, 5, 6}

print(a.union(b))
print(a.intersection(b))
print(a.isdisjoint(b))
print(a.issubset(b))
```

### Dictionary Comprehension

```python
square = {x:x**2 for x in range(1,11) if x % 2 == 0}

# print(list(range(1, 10)))
print(square)

Co_ordinates = [(10,10.5),(20.5,192),(101,102)]
Locations = ["dhaka", "chattogram", "sylhet"]

exact_location = {co_or: loc for co_or, loc in zip(Co_ordinates, Locations)}

print(exact_location)
```

### Problem solving with in Built Structures

```python
# 1. Given a list of numbers, make a list with unique values

lst = [10, 20, 10, 30, 30, 50, 30, 10, 20, 10, 10]

unique_list = list(set(lst))

print(unique_list)
```

```python
# 2. Given a string, print the frequency of the words

string = """A data structure is a specialized format for organizing, processing,
retrieving, and storing data in a computer so that it can be used effectively.
By choosing the right data structure, programmers can significantly lower the time
and space complexity of operations like searching, sorting, insertion, and deletion.
"""

words = string.split()

count = {}

for word in words:
    count[word] = count.get(word, 0) + 1

print(count)

for k, v in count.items():
    print(f"count of {k} is {v}")
```
