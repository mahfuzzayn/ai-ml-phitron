## Problem: Digit Summation

```python
x = int(input())

y = int(input())

last_digit_x = x % 10

last_digit_y = y % 10

sum = last_digit_x + last_digit_y

print(sum)
```

```python
inp = input()

numbers = inp.split()


x = int(numbers[0])

y = int(numbers[1])

last_digit_x = x % 10

last_digit_y = y % 10

sum = last_digit_x + last_digit_y

print(sum)
```

```python
inp = input()

numbers = inp.split()

x = int(numbers[0]); y = int(numbers[1]); z = int(numbers[2])

def getBiggest(x, y, z): 
    if (x > y) and (x > z):
        return x
    elif (y > x) and (y > z):
        return y
    else:
        return z
    
def getSmallest(x, y, z): 
    if (x < y) and (x < z):
        return x
    elif (y < x) and (y < z):
        return y
    else:
        return z
    
biggest = getBiggest(x, y, z)
smallest = getSmallest(x, y, z)

print(smallest, biggest)
```
