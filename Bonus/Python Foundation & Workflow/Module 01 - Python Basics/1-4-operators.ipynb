```python
x = 10; y = 3

sum = x + y # Summation
sub = x - y # Subtraction
mul = x * y # Multiplication
div = round(x / y, 2) # Division

rem = x % y # Remainder

print(sum, sub, mul, div, rem)
```

```python
x = 10; y = 3

greater_than = x > y

greater_than_equal = 10 >= 10

print(greater_than, greater_than_equal)


less_than = x < y

less_than_equal = 10 <= 10

print(less_than, less_than_equal)


equal = x == y

print(equal)

not_equal = x != y

print(not_equal)
```
