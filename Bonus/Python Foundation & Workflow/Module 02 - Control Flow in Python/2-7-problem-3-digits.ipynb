### Digits

```python
number = int(input())

while (number > 0):
    print(number % 10, end = " ")
    # number /= int(number/10)
    number //= 10

```

```python
t = int(input())

for i in range(t):
    number = int(input())
    
    if number == 0:
        print(0)
        continue

    while (number > 0):
        print(number % 10, end = " ")
        # number /= int(number/10)
        number //= 10
    print()
```
