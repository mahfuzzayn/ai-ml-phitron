## Splitting, Joining, Formatted String

```python
prompt = "What is Phitron ?"

tokens = prompt.split()

print(type(tokens)) # list
print(type(tokens[0])) # string
 
sentence = "-".join(tokens)

print(tokens)

print(sentence)

print(type(sentence)) # string

```
