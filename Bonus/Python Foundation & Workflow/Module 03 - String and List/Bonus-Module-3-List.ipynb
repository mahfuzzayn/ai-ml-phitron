## List as Stack

```python
stack = []

# Stack e push
stack.append(1)
stack.append(2)
stack.append(3)
stack.append(4)
stack.append(5)

# Getting the top element
print(stack.pop())
print(stack.pop())

# top element
print(f"Top element: {stack[-1]}")
```

## Queue as List

```python
queue = []

# insertion
queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)

# remove
print(queue.pop(0))

# accessing the front element
print(f"Front e: {queue[0]}")

print(queue.pop(0))
print(f"Front e: {queue[0]}")

print(queue.pop(0))
print(queue.pop(0))
print(queue.pop(0))
```

### List Comprehension

```python
even = []

## Naieve way
for i in range(101):
    if i % 2 == 0: 
        even.append(i)

print(even)
```

```python
random_list = [x + 10 for x in range(1,101) if x % 2 != 0]

print(random_list)
```

### List Comprehension with List of String

```python
fruits = ['apple', 'orange', 'lichi']

capitalize_fruits = []

for fruit in fruits:
    capitalize_fruits.append(fruit.capitalize())

print(capitalize_fruits)
```

```python
mod_capitalize_fruits = [fruit.capitalize() for fruit in fruits]

print(mod_capitalize_fruits)
```
