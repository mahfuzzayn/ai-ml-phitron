### Smart Light Pilot

```python
inp = input()
brightness = float(inp.split()[0])
threshold = float(inp.split()[1])

if brightness >= threshold:
    print("ON")
else:
    print("OFF")
```

### Sensor Calibrator — Min-Max Scaling


```python
inp = input()
numbers = [float(num) for num in inp.split()]

result = (numbers[0] - numbers[1]) / (numbers[2] - numbers[1])

print(f"{result:.2f}")
```

### Mini Loss Tracker


```python
n = int(input())
target = float(input())
numbers = []

for i in range(n):
    numbers.append(float(input()))

average = (sum(numbers)) / n

if (average <= target):
    print("PASS")
else:
    print("RETRY")
```

### Majority Vote Gate


```python
n = int(input())
yes_votes = 0
no_votes = 0

for i in range(n):
    vote = input()

    if vote == "YES":
        yes_votes = yes_votes + 1
    elif vote == "NO":
        no_votes = no_votes + 1

if yes_votes >= no_votes:
    print("ACCEPT")
else:
    print("REJECT")
```

### AI Mood Detector

```python
happy_words = ['happy','joy','smile']
sad_words = ['sad','cry','angry']

sentence = input().lower()
words = sentence.split()
found = False

for i in range(len(words)):
    for j in range(3):
        if (words[i] == happy_words[j]):
            print("Happy Mood")
            found = True
            break
        elif (words[i] == sad_words[j]):
            print("Sad Mood")
            found = True
            break
        
if found != True:
    print("Neutral Mood")

        
```

### AI Brightness Analyzer

```python
inp = input()
numbers = [float(num) for num in inp.split()]
num_of_pixels = len(numbers)
brightness = (sum(numbers) / num_of_pixels)

if (brightness > 170.0):
    print("Bright Image")
elif (brightness >= 85):
    print("Normal Image")
else:
    print("Dark Image")
```

### AI Bias Detector

```python
inp = input()
characters = inp.split()
total_chars = float(len(characters))
total_a = 0.0
total_b = 0.0

for i in range(len(characters)):
    if characters[i] == "A":
        total_a = total_a + 1
    else:
        total_b = total_b + 1

if (total_a / total_chars) > 0.7:
    print("Biased Model")
elif (total_b / total_chars) > 0.7:
    print("Biased Model")
else:
    print("Fair Model")
    
```

### AI Keyword Counter

```python
sentence = input().lower()
words = sentence.split()
ai_keywords = ['ai', 'data', 'model', 'learn', 'train', 'neural']
occurence = 0

for word in words:
    word = word.strip(".,!?;:")
    
    if word in ai_keywords:
        occurence += 1

if occurence >= 2:
    print("AI Detected")
else:
    print("Not AI Related")
            
```
