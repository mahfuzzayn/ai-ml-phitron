### E. Lowest Number


```python
n = input()
inp = input()
numbers = [int(x) for x in inp.split()]
mod_numbers = list(numbers)
mod_numbers.sort()

print(mod_numbers[0], numbers.index(mod_numbers[0]+1))
```

### M. Replace MinMax

```python
n = input()
inp = input()
numbers = [int(x) for x in inp.split()]
mod_numbers = numbers.copy()
mod_numbers.sort()
min_number = mod_numbers.pop(0)
max_number = mod_numbers.pop()
min_number_index = numbers.index(min_number)
max_number_index = numbers.index(max_number)

numbers[min_number_index] = max_number
numbers[max_number_index] = min_number

for i in range(int(n)):
    print(numbers[i], end=" ")
```

### K. Sum Digits

```python
n = input()
inp = input()
numbers = [int(d) for d in inp]
total = 0

for i in range(int(n)):
    total = total + numbers[i]

print(total)
```

### H. Good or Bad

```python
n = int(input())

for i in range(n):
    number = input()
    if number.find("010") == 0 | number.find("101"):
        print("Bad")
    else:
        print("Good")
```
