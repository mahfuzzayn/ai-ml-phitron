### Conditionals statement in Python

```python
raining = True

if raining == True:
    print("Rain coat is necessary")
    print("Caution")
else:
    print("Go without a rain coat")
    print("This is a sunny day")
```

```python
raining = True
Muddy = True

if raining == True:
    print("Rain coat is necessary")
    if Muddy == True:
        print("Sandal is necessary")
    else: 
        print("Bring Sneakers")
else:
    print("Go without a rain coat")
    print("This is a sunny day")
```

```python
taka = 300

## 500 er beshi hole ghurte jabo
## 400 er beshi hole bashai order korbo
## 300 er beshi hole food court e jabo
## er cheye kom thakle savings

if taka > 500:
    print("ghurte jabo")
elif taka > 400:
    print("Bashai order korbo")
elif taka > 300:
    print("Food court e jabo")
else:
    print("Taka ta save korbo")
```

```python
taka = 1000
raining = False

## 500 er beshi and jodi brishti na hoi

if taka > 500 and raining == False:
    print("ghurte jabo")
else:
    print("Taka ta save korbo")
```
