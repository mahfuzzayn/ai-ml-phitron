```python
# Initializing Class

# Base class
class Phone:
    category = "Electronics"
    
    # Constructor
    def __init__(self, model, battery, camera, battery_percentage = 100):
        self.model = model
        self.battery = battery
        self.camera = camera
        self.battery_percentage = battery_percentage
        
    # Methods
    def charge(self, hour):
        print(f"charge completed by {hour}")
        
    def capture(self, photo):
        if (self.battery_percentage) <0:
            print("No charge")
        else:
            self.battery_percentage -= photo
            print(f"Photo captured in {self.model}")
    
    
class Cooling_Mechanism:
    def __init__(self, cooling_method):
        self.cooling_method = cooling_method
        
    # Methods
    def cooling_on(self):
        print(f"The system is being cool by {self.cooling_method}")
```

```python
## Creating Object

apple = Phone("iPhone 17", "3000", "40MP")
apple.category = "Super Electronics"

blueberry = Phone("B-17", 4000, "30MP")

motorola = Phone("M-17", 3500, 35)

apple.capture(10)
motorola.capture(1)

print(apple.battery_percentage)
print(motorola.battery_percentage)
```

### 5-4: OOP Inheritance

### Inheritance: Single

```python
# Child / derived class
class SmartPhone(Phone):
    def __init__(self, model, battery, camera, processor):
        super().__init__(model, battery, camera)
        self.processor = processor
        
    def charge(self, hour):
        print("fast charging in process")
        super().charge(hour)
         
```

```python
pro = SmartPhone("X", 5000, 48, "SnapDragon")

print(pro.model)

pro.charge(1)
```

### Inheritance: Multiple

```python
class SmartPhone_CoolingMode(SmartPhone, Cooling_Mechanism):
    def __init__(self, model, battery, camera, processor, cooling_method):
        SmartPhone.__init__(self, model, battery, camera, processor)
        Cooling_Mechanism.__init__(self, cooling_method)
```

```python
pro_cooling = SmartPhone_CoolingMode("Y", 5000, 100, "SD", "Nitrogen")

print(pro_cooling.processor) # SmartPhone class theke
print(pro_cooling.battery) # Phone class teke
print(pro_cooling.cooling_method()) # Cooling_Mechanism class teke
print(pro_cooling.cooling_on()) # Cooling_Mechanism class er method
print(pro_cooling.charge(10)) # SmartPhone er modified charge jeta Phone class theke inherited
```

```python
# Practice

class Car:
    def __init__(self, brand, model, topSpeed, mileage):
        self.brand = brand
        self.model = model
        self.topSpeed = topSpeed
        self.mileage = mileage
        
    def charge(self, hour):
        print(f"Started charging for {hour} hours")
        
        
class Car_Cooling:
    def __init__(self, cooling_method):
        self.cooling_method = cooling_method
    
    def charge(self, hour):
        print(f"Car is charing for {hour} hours with Cooling mode on")
        
class Customer_Car(Car, Car_Cooling):
    def __init__(self, brand, model, topSpeed, mileage, cooling_method):
        Car.__init__(self, brand, model, topSpeed, mileage)
        Car_Cooling.__init__(self, cooling_method)
        
tesla = Customer_Car("Tesla", "X", 200, 100, "Heated")

tesla.charge(10)
```

### 5-5: OOP Polymorphism

```python
class Camera:
    def __init__(self, name):
        self.name = name
        
    def capture(self):
        print("A photo is captured")
```

```python
class Smart_Phone(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution
        
    # Method Override
    def capture(self):
        print("Photo is captured by a phone")
        
class DSLR(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution
        
    # Method Override
    def capture(self):
        print("Photo is captured by a DSLR")
        
class Drone(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution
    
    # Method Override
    def capture(self):
        print("Photo is captured by a Drone")
        
phone = Smart_Phone("Phone", 30)
dslr = DSLR("DSLR", 200)
drone = Drone("Drone", 150)

phone.capture()
dslr.capture()
drone.capture()
        
```

### 5-6: OOP Encapsulation

```python
# Design
class Mobile:
    def __init__(self, name, model, imei):
        self.__name = name
        self.__model = model
        self.__imei = imei # Making it private
     
    def charge():
        print("Phone is charging")   
        
    # Method to get imei
    def imei_getter(self):
        return self.__imei
        
    def model_getter(self):
        return self.__model   

    def name_getter(self):
        return self.__name
    
    def name_setter(self, name):
        self.__name = name  

# Outside World

iPhone = Mobile("Phone", "17", "1xkaf1")

print(iPhone.name_getter())
print(iPhone.model_getter())
print(iPhone.imei_getter())

iPhone.name = "Ph"
iPhone.imei = "123"

iPhone.name_setter("Apple")
print(iPhone.name_getter())
print(iPhone.imei_getter())
```
