```python
### 2-4 Positional Arguments
```

```python
def adder(a,b,c):
    print(f"{a},{b},{c}")

adder(10,30,20)
```

```python
def adder(*args):
    print(args)
    
adder(10, 30, 20, 40)
```

```python
### 2-5 Keyword Arguments
```

```python
def adder(a = 10, b = 30, c = 20):
    print(f"{a},{b},{c}")
    
adder(b=20,c=30,a=10)
```

```python
def adder(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

adder(a = 10, b = 30, c = 40, n_estimator = 50)
```

```python
### 2-6 Return Type of a Function
```

```python
def adder(a, b):
    sum = a + b
    return sum

print(type(adder(10, 20)))
```

```python
def adder(a, b):
    sum = a + b
    return sum

print(type(adder(10.5, 20)))
```

```python
def welcome_message():
    return "Hello user"

print(type(welcome_message()))
```

```python
def adder(a, b):
    return a, b # Tuple

print(type(adder(10.5, 20)))

# unzip

x, y = adder(10.5, 20)

print(x, y)
```

```python
def adder(a, b):
    return [a, b] # List

print(type(adder(10.5, 20)))
```

```python
def adder(a, b):
    return {a, b} # Set

print(type(adder(10.5, 20)))
```

### Generator

```python
lst = [ x for x in range(500) ] # dataset

def data_loader(chunk_size, lst):
    for i in range(0,len(lst), chunk_size):
        yield lst[i:i+chunk_size]

x = data_loader(10, lst)

print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
```
