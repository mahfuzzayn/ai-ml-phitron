### 3-5 Practice Problems

Practice 1 :
Get the text from sample.txt , and count the following

Number of lines

Number of words

Number of characters

and save it to the counter_of_string.txt file

```python
with open("./sample_data/sample.txt", "r") as file:
    strings = file.readlines()
    file.seek(0)
    words = file.read()
    file.seek(0)
    characters = list(words)
    
    print(f"Total lines: {len(strings)}")
    print(f"Total words: {len(words.split())}")
    print(f"Total characters: {len(characters)}")
    
    print("Storing on counter_of_string.txt file")
    
    with open("./sample_data/counter_of_string.txt", "w") as file:
        file.write(f"Total lines: {len(strings)}\n")
        file.write(f"Total words: {len(words.split())}\n")
        file.write(f"Total characters: {len(characters)}\n")
    
```

```python
from functools import reduce

with open("./sample_data/sample.txt", "r") as file:
    string_list = file.readlines()
    
    # Total lines
    total_lines = len(string_list)
    
    # Total words
    number_of_words = list(map(lambda x: len(x.split()), string_list))
    total_number_of_words = reduce(lambda x, y : x + y, number_of_words)
    
    # Cleaning process
    
    # 1. new line delete
    string_list = list(map(str.strip,string_list))
    

    
    # 2. space delete
    string_list = list(map(lambda x: x.replace(" ", ""), string_list))
    number_of_character = list(map(lambda x: len(x), string_list))
    total_number_of_characters = reduce(lambda x, y: x + y, number_of_character)
    
    print(f"Total lines: {total_lines}")
    print(f"Total words: {total_number_of_words}")
    print(f"Total characters: {total_number_of_characters}")
    
    with open("./sample_data/counter_of_string.txt", "w") as file:
        file.write(f"Total lines: {total_lines}\nTotal words: {total_number_of_words}\nTotal characters: {total_number_of_characters}\n")
        
```

### 3-6 Exceptions and why handling them

```python
# Zero Division Error

n = int(input())

try:
    a = 10/n
except ZeroDivisionError:
    print("you can't divide a number by 0")

print("hello world")
```

```python
# Name Error

try:
    x = y
except NameError as ne:
    print(ne)

print("hello world")
```

```python
# Unknown Random Exception

try:
    x = y
except Exception as ne:
    print(ne)

print("hello world")
```
