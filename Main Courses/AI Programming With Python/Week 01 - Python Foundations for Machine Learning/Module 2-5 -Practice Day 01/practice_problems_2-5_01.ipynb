### Problem 2

```python
def batch_generator(data, batch_size):
    for i in range(0, len(data), batch_size):
        yield data[i:i+batch_size]
    

data = [1, 2, 3, 4, 5, 6, 7, 8]
batch_size = 3

x = batch_generator(data, batch_size)

print(next(x))
print(next(x))
print(next(x))
```

### Problem 4

```python
def power(base, exponent=0):
    if exponent != 0:
        return base ** exponent
    else:
        return base * base
    
power(5)
```

### Problem 5 (Conceptual Session)

```python
data = [1, 2, 3, 4, 5, 6]

even_numbers = list(filter(lambda x : x % 2 == 0, data))

squared_even_numbers = list(map(lambda x: x * x, even_numbers))


print(squared_even_numbers)
```
