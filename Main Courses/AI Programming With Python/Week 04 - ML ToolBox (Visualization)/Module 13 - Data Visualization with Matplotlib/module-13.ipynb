### 13-2: 2D Line Plot Basics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hours = [2, 3, 4, 1, 5, 7, 3]
days = [1, 2, 3, 4, 5, 6, 7]

plt.plot(days, hours)
```

### 13-3: 2D Line from Pandas DF

```python
df = pd.read_csv("./enrollment_data.csv")

df
```

```python
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title("Student Enrollment over years")

plt.plot(df['Year'], df['Programming'])
plt.plot(df['Year'], df['Digital Marketing'])
```

```python
# Adding labels using plt.legend()

plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title("Student Enrollment over years")

plt.plot(df['Year'], df['Programming'], label = 'Programming')
plt.plot(df['Year'], df['Digital Marketing'], label = 'Digital Marketing')
plt.legend(loc="upper left") # upper right, lower left, best
```

### 13-4: Customzing Line plot (color, linestyle, marker) and Grid

```python
# Adding some visulization

plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title("Student Enrollment over years")

plt.plot(df['Year'], df['Programming'], label = 'Programming', color="#3498db", linewidth=3, linestyle="dashed", marker="o", markersize=7)
plt.plot(df['Year'], df['Digital Marketing'], label = 'Digital Marketing', color="crimson", linewidth=3, linestyle="dotted")
plt.legend()

```

```python
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title("Student Enrollment over years")

plt.plot(df['Year'], df['Programming'], label = 'Programming', color="#3498db", linewidth=3, linestyle="dashed", marker="o", markersize=7)
plt.plot(df['Year'], df['Digital Marketing'], label = 'Digital Marketing', color="crimson", linewidth=3, linestyle="dotted")
plt.legend()
plt.grid()

plt.show() # Run this too see the output of the plot
```

### 13-6: Scatter Plots

```python
students = pd.read_csv("./student-IQdata.csv")

students
```

```python
# Relation between Study Hours and IQ Score -> Positive Correlation

plt.xlabel("Study Hours")
plt.ylabel("IQ Score")
plt.title("Relation between Study Hours and IQ Score")
plt.scatter(students['Study_Hour'], students['IQ_Score'])
```

```python
# Relation between Shoe Size and IQ Score -> No Correlation

plt.xlabel("Shoe Size")
plt.ylabel("IQ Score")
plt.title("Relation between Shoe Size and IQ Score")
plt.scatter(students['Shoe_Size'], students['IQ_Score'])
```

```python
# Relation between Chilling Hours and IQ Score -> Negative Correlation

plt.xlabel("Chilling Hours")
plt.ylabel("IQ Score")
plt.title("Relation between Chilling Hours and IQ Score")
plt.scatter(students['Chilling_Hours'], students['IQ_Score'], marker='o', color="#3498db", label="Chilling Score and IQ Score") # Also designings can be done: Color, Size, Marker etc.
plt.legend()
```

### 13-7: Histogram

### 13-8: Bar Chart

```python
data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)

df
```

```python
# Gender wise

gender_group = df.groupby("Gender").size()

display(gender_group.index)
display(gender_group.values)
```

```python
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Bar Chart of Gender among Students")
plt.bar(gender_group.index, gender_group.values, color=["#3498db", "red"], edgecolor="black")
```

```python
df = pd.read_csv("./student_data.csv")

df
```

```python
# Course Completion

status = df.groupby("CompletionStatus").size()

plt.xlabel("Completion Status")
plt.ylabel("Count")
plt.title("Bar Chart of Course completion")
plt.bar(status.index, status.values, color=['green', "orange", "red"], edgecolor="black")
plt.show()
```

```python
# Location

location = df.groupby("Location").size()

plt.xlabel("Location")
plt.ylabel("Count")
plt.title("Bar Chart of Location of Students")
plt.bar(location.index, location.values, edgecolor="black")
plt.show()
```

### 13-9: Pie Chart

```python
# Course Completion Status

status = df.groupby("CompletionStatus").size()

plt.pie(status, labels=status.index, autopct='%1.1f%%', explode=(0.1, 0, 0), shadow=True)
plt.show()
```

```python
# Location

location = df.groupby("Location").size()

plt.pie(location, labels=location.index, autopct='%1.1f%%')
plt.show()
```

```python
# Instructor

instructor = df.groupby("Instructor").size()

plt.pie(instructor, labels=instructor.index, autopct='%1.1f%%')
plt.show()
```
