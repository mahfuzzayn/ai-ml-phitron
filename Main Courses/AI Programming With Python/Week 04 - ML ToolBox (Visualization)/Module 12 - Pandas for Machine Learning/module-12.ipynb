### 12-1: Filtering Based on String

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah","Sakib"],
    "City": ["New York", "Los Angeles", "Newark", "Boston", "New Delhi", "Chicago", "New Orleans", "Houston","H Los Ang"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Marketing", "Finance", "HR", "HR"],
    "Salary": [50000, 60000, 55000, 70000, 52000, 58000, 62000, 51000,70000]
}

df = pd.DataFrame(data)

df
```

```python
df.loc[df['City'].str.contains("New")] # Case sensitive
```

```python
df.loc[df['City'].str.contains("new", case=False)] # Case insensitive

# We can also use regular expression

# Starts with Los
df.loc[df['City'].str.contains(r"^Los")]
```

```python
df.loc[df['City'].str.contains(r"Los")] 
```

```python
# Ends with rk

df.loc[df['City'].str.contains(r"rk$")] 
```

```python
# Name starts with a vowel

df.loc[df['Name'].str.contains(r"^[AEIOU]")] 
```

```python
# String contains New or Los

df.loc[df['City'].str.contains(r"New|Los")] 
```

### 12-7: Apply Function in DF

```python
# Min max Scaling

mn = df['Total Marks'].min()
mx = df['Total Marks'].max()

# Scaled = (x-min) / (mx-min)

df['Scaled Marks'] = df['Total Marks'].apply(lambda x: (x-mn)/(mx-mn))

df
```

```python
# Custom built function

def grading_system(marks):
    if marks >= 260:
        return "A+"
    elif marks >= 250:
        return "A"
    else:
        return "A-"
    
df['Grade'] = df['Total Marks'].apply(grading_system)

df
```

```python
def marking_system(df):
    a = df['Data Structure Marks'] * 3
    b = df['Python Marks'] * 3
    c = df['Algorithm Marks'] * 4

    return a + b + c

df['Exceptional Marks'] = df.apply(marking_system, axis=1) # Row wise that's why

df
```

### 12-8: DateTime and Timedelta

```python
df['EnrollmentDate']
```

```python
# We are converting into Pandas Datetime64 so that we can work around with rows with timing conditions
# Otherwise we have to do manual splitting from str and then work

df['EnrollmentDate'] = pd.to_datetime(df['EnrollmentDate'])
```

```python
# Creating new columns from that EnrollmentDate
df['Enrollment Year'] = df['EnrollmentDate'].dt.year

df['Enrollment Date'] = df['EnrollmentDate'].dt.day

df
```

```python
# Time Delta

df['FinishedDate'] = pd.to_datetime(df['FinishedDate'])

df['Total time taken to finish'] = df['FinishedDate'] - df['EnrollmentDate']

df
```

### 12-9: Pandas Group By

```python
instructor_group = df.groupby("Instructor")

instructor_group
```
