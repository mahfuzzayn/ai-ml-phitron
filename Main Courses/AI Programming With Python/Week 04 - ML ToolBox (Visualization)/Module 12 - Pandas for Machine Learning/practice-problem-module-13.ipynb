### Problem 1

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})
                   
print(df)  

# Replace null values based on given conditions
display(df.fillna({'A': 0.0, 'B': 99.0, 'C': 'Unknown'}))


# Forward fill column B
df2 = df.copy()
df2['B'] = df2['B'].ffill()
display(df2)


# Fills missing values in numeric columns with the mean of the column and in string columns with 'Unknown'.
df3 = df.copy()

num_cols = df3.select_dtypes(include=[np.number]).columns
df3[num_cols] = df3[num_cols].fillna(df3[num_cols].mean())

str_cols = df3.select_dtypes(include=['object', 'category']).columns
df3[str_cols] = df3[str_cols].fillna('Unknown')

df3





```

### Problem 2

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})
                   
display(df)

# Drop rows where column A has missing values
display(df.dropna(subset=['A']))


# Drop rows with less than 2 non-null values
display(df.dropna(thresh=2))


# Drop rows where both Column A and Column B has missing values
display(df.dropna(subset=['A', 'B'], how='all'))

```

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
df = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Value1': [10, 20, 30, 40, 50, 60],
    'Value2': [100, 200, 300, 400, 500, 600]
})

# Basic groupby and aggregation
print("Mean of Value1 for each Category:")
print(df.groupby('Category')['Value1'].mean())

print("Max of Value2 for each Category:")
print(df.groupby('Category')['Value2'].max())

print("Sum of Value1 for each Category:")
print(df.groupby('Category')['Value1'].sum())
```
