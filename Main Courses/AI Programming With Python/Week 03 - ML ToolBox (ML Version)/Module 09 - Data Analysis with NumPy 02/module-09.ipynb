##### 9-1: ndarray Manipulation -1

```python
import numpy as np

arr = np.random.randint(1, 100, size=(10, 5)) # Sample - 10, feature 5

print(arr.shape)
```

```python
# Changing Shape

b = arr.reshape(5, 10) # All data must exists can't say (3, 2) on 50 items array 

print(b.shape)
print(b.ndim)
print(b)
```

```python
flatten = b.flatten() # to convert to 1d array

print(flatten)
print(flatten.ndim)

column_wise_flattening = np.ravel(b, order="F") # Column Wise
row_wise_flattening = np.ravel(b, order="C") # Column Wise

print(column_wise_flattening)
print(row_wise_flattening)
```

```python
# Array Concatenation

a = np.random.randint(1, 10, size=(2, 3))
b = np.random.randint(20, 30, size=(2, 3))

print(a)
print(b)

# We want to merge two array and then train model
# Two ways: Row & Column wise
# Row wise concatenate: Column count should match
# Column wise concatenate: Row count should match

con_row_wise = np.concatenate((a, b), axis=0) # Row wise
con_col_wise = np.concatenate((a, b), axis=1) # Column wise

print(con_row_wise)
print(con_col_wise)
```

### 9-2: ndarray Manipulation - 2

```python
# Transpose

mat = np.array([[10, 20, 30],
               [30, 40, 50]
               ])

transpose = mat.T

print(transpose)
```

```python
# Array Split

a = np.random.randint(1, 10, size=(10, ))

# Array Split -> Dynamically adjust items
splitted_array = np.array_split(a, 3)

print(a)
print(splitted_array)

# Split -> Equal Division
splitted_array = np.split(a, 3)
```

### 9-3: ndarray Arithmetic Operators and Mathematical function

```python
# Faster for vectorization

x = np.array([10, 8, 20, 30, 40, 100])
y = np.array([2, 3, 4, 5])

add = x + y # Addition
add_2 = np.add(x, y)
print(add)
print(add_2)
# In tradition list we can't do this instead it will add items

sub = x - y
print(sub)

mul = x * y
print(mul)

div = x / y
print(div)

remainder = x % y
print(remainder)

# Also np.subtract(), multiply(), divide(), remainder()

```

```python
# Mathematical Function

x = np.array([10, 8, 20, 30, 16, 40, 100])

# Trigonometry
# sin, cos, tan values in radian
sin_val = np.sin(x) 
print(sin_val)

cos_val = np.cos(x)
print(cos_val)

tan_val = np.tan(x)
print(tan_val)

# Degree: D * PI / 180
# Radian: R * 180 / PI
deg_con = np.rad2deg(x)
print(deg_con)

ran_con = np.deg2rad(deg_con)
print(deg_con)

# Log

base_10_log_val = np.log10(x)
print(base_10_log_val)

base_2_log_val = np.log2(x)
print(base_2_log_val)

# Square Root

sqrt = np.sqrt(x)

print(sqrt)
```

```python
# Different

s = np.sum(x)
print(s)

# Prefix sum
cuml = np.cumulative_sum(x)
print(cuml)
```
