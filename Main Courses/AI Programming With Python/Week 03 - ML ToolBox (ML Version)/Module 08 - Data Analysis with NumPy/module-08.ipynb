### Ndarray

```python
import numpy as np # NumPy Package

# 1d array

arr1 = np.array([1, 2, 3, 4, 5])

# print(type(arr1))
# print(arr1.ndim) # Dimension

# 2d array

arr2 = np.array([[1, 2, 3],  # row 1
                 [4, 5, 6]]) # row 2

print(arr2.ndim)

# 3d array

arr3 = np.array([
                # 1st floor
                [[1, 2, 3], # row 1
                 [4, 5, 6]], # row 2
                
                # 2nd floor
                [[7, 8, 9], # row 1
                 [10, 11, 12]], # row 2
                
                # 3rd floor
                [[7, 8, 9], # row 1
                 [10, 11, 12]] # row 2
                ])


print(arr3.ndim)
```

### Array attributes

```python
# dimension

print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

# shape

print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

# data type

print(arr1.dtype)
print(arr2.dtype)
print(arr1.dtype)

# size

print(arr1.size)
print(arr2.size)
print(arr3.size)
```

```python
arr = np.array([1, 2, 3])

print(arr.dtype)

arr = np.array([1.5, 1.6, 1.7])

arr = np.array([1.5, 1.6, '3'])

print(arr.dtype)
```

```python
# Upcasted to float
arr = np.array([1, 2, 3.2])

# Upcasted to string
arr = np.array([1.2, 3.2, 'hello'])

print(arr.dtype)
```
